#### Ques 5 : 	Convert the dog types to one column (Issue 1 in the tidiness issues)

Issue 1.	a lot of dogs don't have an age category (floofer,doggo,pupper,puppo)

ANS : There's no column named Age. Hence, Combining four columns into one

```python
tweets[["doggo","floofer","pupper","puppo"]]=tweets[["doggo","floofer","pupper","puppo"]].replace("None","")
```

```python
tweets['dog_type']=tweets[["doggo","floofer","pupper","puppo"]].apply(lambda x: ''.join(x), axis=1)
```
